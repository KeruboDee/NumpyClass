```python
import numpy as np
```

```python
# y = mx+b
# datapoints to be used
x = [1, 2, 3, 4, 5]
y = [5, 4, 6, 5, 6]
```

```python
# create a numpy array with floats
arrx = np.array(x, dtype = np.float64)
arry = np.array(y, dtype = np.float64)
arrx
arry
```

```python
# to find m which is the gradient/ slope 
m = round((((np.mean(arrx) * np.mean(arry)) - np.mean(arrx * arry))/
   (((np.mean(arrx))**2) - (np.mean(arrx**2)))), 1)
m
```

```python
# to find the best fit, b, which is also the (y-intercept)
b = (np.mean(arry) - (m * (np.mean(arrx))))
b    
```

```python
# to find the regression-line
regression_line = []
for i in arrx:
    regression_line.append((m * i) + b)
    
regression_line
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
plt.scatter(arrx, arry, color = '#003F72')
plt.plot(arrx, regression_line)
plt.show()
```

```python
# to predict a couple of points with our input data
predict_x = 7
predict_y = (m * predict_x) + b

plt.scatter(arrx, arry, color = '#003F72', label = "Data")
plt.plot(arrx, regression_line, label = "Regression Line")
plt.legend(loc = 4)
plt.show()
```
